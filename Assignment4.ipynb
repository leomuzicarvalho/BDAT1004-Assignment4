```python
def frequencies(text):
    letters = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    numbers = [0 for x in range(len(letters))]
    for i in range(len(letters)):
        for j in text.lower():
            if(letters[i] == j):
                numbers[i]+=1
    print(numbers)

text = input("Type any text: ")
frequencies(text)
```

Question 6 (Leoanrdo did)

```python
def sieve(n):
    PrimeL = []
    L = [i for i in range(2,n)]
    while(len(L)>0):
        number = L.pop(0)
        PrimeL.append(number)
        lengthL = len(L)
        for j in range(lengthL):
            if(j>=len(L)):
                break
            if (L[j]%number==0):
                L.pop(j)
    return PrimeL

print(sieve(368))
```

```python
import re

def wordCount():
    wordList = open('frankenstein.txt').read().split()
    
    counters = {}
    for word in wordList:
        strippedWord = re.sub('[^A-Za-z0-9]+', '', word).lower()
        if(len(strippedWord)<=2):
            continue
        if strippedWord in counters:
            counters[strippedWord] += 1
        else:
            counters[strippedWord] = 1
    
    for strippedWord in counters:
        if counters[strippedWord] == 1:
            print('{:8} appears {} time.'.format(strippedWord, counters[strippedWord]))
        else:
            print('{:8} appears {} times.'.format(strippedWord, counters[strippedWord]))

wordCount()

```
